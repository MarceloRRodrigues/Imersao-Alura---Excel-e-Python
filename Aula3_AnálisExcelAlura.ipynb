```python
import pandas as pd
import plotly.express as px
```

```python
df_variação_mensal = pd.read_excel("/content/Exercício Alura - Tabela de ações (1).xlsx", sheet_name="variação mensal")
df_variação_mensal
```

```python
df_total_de_ações = pd.read_excel("/content/Exercício Alura - Tabela de ações (1).xlsx", sheet_name="Total_de_acoes")
df_total_de_ações
```

```python
df_ticker = pd.read_excel("/content/Exercício Alura - Tabela de ações (1).xlsx", sheet_name="Ticker")
df_ticker
```

```python
df_variação_mensal = df_variação_mensal[['Ativo', 'Data','Último (R$)','Var. Mês (%)']].copy()
df_variação_mensal
```

```python
df_variação_mensal = df_variação_mensal.rename(columns={'Último (R$)':'valor_final','Var. Mês (%)':'var_mes_pct'}).copy()
df_variação_mensal
```

```python
df_variação_mensal['var_pct'] = df_variação_mensal['var_mes_pct']/100
df_variação_mensal
```

```python
df_variação_mensal['valor_inicial'] = df_variação_mensal['valor_final']/(df_variação_mensal['var_pct']+1)
df_variação_mensal
```

```python
df_variação_mensal = df_variação_mensal.merge(df_total_de_ações, left_on='Ativo', right_on='Código', how='left')
df_variação_mensal
```

```python
df_variação_mensal= df_variação_mensal.drop(columns='Código')
df_variação_mensal
```

```python
df_variação_mensal['variacao_rs'] = (df_variação_mensal['valor_final']-df_variação_mensal['valor_inicial'])*df_variação_mensal['Qtde. Teórica']
df_variação_mensal
```

```python
df_variação_mensal = df_variação_mensal.rename(columns={'Qtde. Teórica':'qtd_acoes'}).copy()
df_variação_mensal
```

```python
pd.options.display.float_format = '{:.2f}'.format
df_variação_mensal
```

```python
df_variação_mensal['qtd_acoes'] = df_variação_mensal['qtd_acoes'].astype(int)
df_variação_mensal
```

```python
df_variação_mensal['Resultado'] = df_variação_mensal['variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_variação_mensal
```

```python
df_variação_mensal = df_variação_mensal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_variação_mensal
```

```python

df_variação_mensal
```

```python
df_variação_mensal['Segmento']= df_variação_mensal['Segmento'].copy()
df_variação_mensal
```

```python
df_variação_mensal = df_variação_mensal.rename(columns={'Unnamed: 4':'cat_idade'}).copy()
df_variação_mensal
```

```python
df_variação_mensal= df_variação_mensal.drop(columns='Ticker')
df_variação_mensal
```

```python
df_variação_mensal= df_variação_mensal.drop(columns='cat_idade')
df_variação_mensal
```

```python
df_variação_mensal['Cat_idade'] = df_variação_mensal['idade'].apply(lambda x: 'mais de 100' if x > 100 else ('entre 50 e 100' if x >= 50 else 'menos de 50'))
df_variação_mensal
```

```python
#maior
maior = df_variação_mensal['variacao_rs'].max()

#menor
menor = df_variação_mensal['variacao_rs'].min()

#média
media = df_variação_mensal['variacao_rs'].mean()

#Calculando a média de quem subiu
media_subiu = df_variação_mensal[df_variação_mensal['Resultado'] == 'Subiu']['variacao_rs'].mean()

#Calculando a média de quem desceu
media_desceu = df_variação_mensal[df_variação_mensal['Resultado'] == 'Desceu']['variacao_rs'].mean()

#imprimindo
print(f'Maior\tR${maior:.2f}')

print(f'Menor\tR${menor:.2f}')

print(f'Média\tR${media:.2f}')

print(f'Média de quem subiu\tR${media_subiu:.2f}')

print(f'Média de quem desceu\tR${media_desceu:.2f}')
```

```python
df_variação_mensal_subiu = df_variação_mensal[df_variação_mensal['Resultado'] == 'Subiu']
df_variação_mensal_subiu
```

```python
df_analise_segmento = df_variação_mensal_subiu.groupby('Segmento')['variacao_rs'].sum().reset_index()
pd.options.display.float_format = 'R$ {:,.2f}'.format
df_analise_segmento
```

```python
df_analise_saldo= df_variação_mensal.groupby('Resultado')['variacao_rs'].sum().reset_index()
pd.options.display.float_format = 'R$ {:,.2f}'.format
df_analise_saldo
```

```python

fig = px.bar(df_analise_saldo, x='Resultado', y='variacao_rs', text='variacao_rs', title='Variação R$ x Resultado')
pd.options.display.float_format = 'R$ {:,.2f}'.format
fig.show()
```

```python
fig2 = px.pie(df_analise_segmento, values='variacao_rs', names='Segmento', title='Variação x Segmento (Subiu)')
fig2
```

```python
df_analise_cat_idade = df_variação_mensal.groupby('Cat_idade')['variacao_rs'].sum().reset_index()
pd.options.display.float_format = 'R$ {:,.2f}'.format
df_analise_cat_idade
```

```python
pd.options.display.float_format = 'R$ {:,.2f}'.format
fig3 = px.bar(df_analise_cat_idade, x='Cat_idade', y='variacao_rs', text='variacao_rs', title='Variação R$ x Faixa etária')
fig3.show()
```
